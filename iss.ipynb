```python
import requests
import time

isslocations = []
url = "http://api.open-notify.org/iss-now.json"

for i in range (100):
    response = requests.get(url)
    data = response.json()
    isslocations.append((data["iss_position"]["latitude"], data["iss_position"]["longitude"], data["timestamp"]))
    time.sleep(10)
```

```python
import matplotlib.pyplot as plt

latitude = [location[0] for location in isslocations]
longitude = [location[1] for location in isslocations]

plt.plot(longitude, latitude)

plt.xlabel("Longitude")
plt.ylabel("Latitude")
plt.title("Path of the ISS")

plt.xticks([i * 11 for i in range(11)])
plt.yticks([i * 11 for i in range(11)])
plt.tick_params(labelsize=7)

plt.show()
```
